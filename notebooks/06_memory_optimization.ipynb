```python
import pandas as pd
import numpy as np
```

```python
print("Loading data...")
model_data = pd.read_parquet("../data/processed/preprocessed_data.parquet")
```

```python
print("Optimizing memory...")
start_mem = model_data.memory_usage().sum() / 1024**2
print(start_mem)
```

```python
for col in model_data.columns:
    if model_data[col].dtype == 'int64':
        # if it only contains 0 or 1 we will crush it into int8
        if model_data[col].min() >= 0 and model_data[col].max() <= 1:
            model_data[col] = model_data[col].astype('int8')
        # otherwise, if its a standard integer, int32 is usually plenty
        elif model_data[col].min() >= np.iinfo(np.int32).min and model_data[col].max() <= np.iinfo(np.int32).max:
            model_data[col] = model_data[col].astype('int32')
```

```python
end_mem = model_data.memory_usage().sum() / 1024**2
print(f"Memory after memory optmization: {end_mem:.2f} MB")
```

```python
# Save the memory-optimized dataset
save_path = "../data/processed/preprocessed_data_optimized.parquet"

print(f"Saving optimized dataset to {save_path}...")
model_data.to_parquet(save_path, index=False)
print("Saved successfully!")
```
